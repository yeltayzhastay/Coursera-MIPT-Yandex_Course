### Импортируем необходимые библиотеки

```python
%pylab inline
import pandas as pd
from scipy import stats
import statsmodels.api as sm
import matplotlib.pyplot as plt
import warnings
from itertools import product

def invboxcox(y,lmbda):
    if lmbda == 0:
        return(np.exp(y))
    else:
        return(np.exp(np.log(lmbda*y+1)/lmbda))
```

Считываем данные:

```python
wages = pd.read_csv('WAG_C_M.csv',';', index_col=['month'], parse_dates=['month'], dayfirst=True)
wages.columns = ['wage']
plt.figure(figsize(15,7))
wages.plot()
plt.ylabel('Wages')
pylab.show()
```

Проверка стационарности и STL-декомпозиция ряда:

```python
plt.figure(figsize(15,10))
sm.tsa.seasonal_decompose(wages.wage).plot()
print(f"Критерий Дики-Фуллера: p={sm.tsa.stattools.adfuller(wages.wage)[1]}")
```

Видим, что дисперсия меняется со временем. Проведем её стабилизацию:

```python
wages['wage_box'], lmbda = stats.boxcox(wages.wage)
plt.figure(figsize(15,7))
wages.wage_box.plot()
plt.ylabel(u'Transformed wine sales')
print("Оптимальный параметр преобразования Бокса-Кокса: %f" % lmbda)
print(f"Критерий Дики-Фуллера: p={sm.tsa.stattools.adfuller(wages.wage_box)[1]}")
```

Критерий Дики-Фуллера не отвергает гипотезу нестационарности. Попробуем сезонное дифференцирование; сделаем на продифференцированном ряде STL-декомпозицию и проверим стационарность:

```python
wages['wage_box_diff1'] = wages.wage_box - wages.wage_box.shift(12)
plt.figure(figsize(15,10))
sm.tsa.seasonal_decompose(wages.wage_box_diff1[12:]).plot()   
print("Критерий Дики-Фуллера: p=%f" % sm.tsa.stattools.adfuller(wages.wage_box_diff1[12:])[1])
```

Критерий Дики-Фуллера отвергает гипотезу нестационарности, но визуально в данных виден тренд. Попробуем добавить ещё обычное дифференцирование:

```python
wages['wage_box_diff2'] = wages.wage_box_diff1 - wages.wage_box_diff1.shift(1)
plt.figure(figsize(15,10))
sm.tsa.seasonal_decompose(wages.wage_box_diff2[13:]).plot()   
print(f"Критерий Дики-Фуллера: p={sm.tsa.stattools.adfuller(wages.wage_box_diff2[13:])[1]}")
```

Гипотеза нестационарности уверенно отвергается, и визуально ряд выглядит лучше. 

### Подбор модели
Посмотрим на ACF и PACF полученного ряда:

```python
plt.figure(figsize(15,8))
ax = plt.subplot(211)
sm.graphics.tsa.plot_acf(wages.wage_box_diff2[13:].values.squeeze(), lags=48, ax=ax)
pylab.show()
ax = plt.subplot(212)
sm.graphics.tsa.plot_pacf(wages.wage_box_diff2[13:].values.squeeze(), lags=48, ax=ax)
pylab.show()
```

```python
#сразу расширим область подбираемых значений
ps = range(0, 3)
d=1
qs = range(0, 3)
Ps = range(0, 2)
D=1
Qs = range(0, 2)
```

```python
parameters = product(ps, qs, Ps, Qs)
parameters_list = list(parameters)
len(parameters_list)
```

```python
%%time
results = []
best_aic = float("inf")
warnings.filterwarnings('ignore')

for param in parameters_list:
    #try except нужен, потому что на некоторых наборах параметров модель не обучается
    try:
        model=sm.tsa.statespace.SARIMAX(wages.wage_box, order=(param[0], d, param[1]), 
                                        seasonal_order=(param[2], D, param[3], 12)).fit(disp=-1)
    #выводим параметры, на которых модель не обучается и переходим к следующему набору
    except ValueError:
        print('wrong parameters:', param)
        continue
    aic = model.aic
    #сохраняем лучшую модель, aic, параметры
    if aic < best_aic:
        best_model = model
        best_aic = aic
        best_param = param
    results.append([param, model.aic])
    
warnings.filterwarnings('default')
```

```python
result_table = pd.DataFrame(results)
result_table.columns = ['parameters', 'aic']
print(result_table.sort_values(by = 'aic', ascending=True).head())
```

Лучшая модель:

```python
print(best_model.summary())
```

Её остатки:

```python
plt.figure(figsize(15,8))
plt.subplot(211)
best_model.resid[13:].plot()
plt.ylabel(u'Residuals')

ax = plt.subplot(212)
sm.graphics.tsa.plot_acf(best_model.resid[13:].values.squeeze(), lags=48, ax=ax)

#Смещаем весь ряд на значение среднего, чтобы ряд стал несмещенным.
best_model.resid[13:] = best_model.resid[13:] + best_model.resid[13:].mean()
print("Критерий Стьюдента: p=%f" % stats.ttest_1samp(best_model.resid[13:], 0)[1])
print("Критерий Дики-Фуллера: p=%f" % sm.tsa.stattools.adfuller(best_model.resid[13:])[1])
```

Остатки несмещены (подтверждается критерием Стьюдента) стационарны (подтверждается критерием Дики-Фуллера и визуально), неавтокоррелированы (подтверждается коррелограммой).
Посмотрим, насколько хорошо модель описывает данные:

```python
wages['model'] = invboxcox(best_model.fittedvalues, lmbda)
plt.figure(figsize(15,7))
wages.wage.plot()
wages.model[13:].plot(color='r')
plt.ylabel('Wages')
pylab.show()
```
