```python
def get_bootstrap_samples(data, n_samples):
    indices = np.random.randint(0, len(data), (n_samples, len(data)))
    samples = data[indices]
    return samples

def stat_intervals(stat, alpha):
    boundaries = np.percentile(stat, [100 * alpha / 2., 100 * (1 - alpha / 2.)])
    return boundaries
```

```python
np.random.seed(0)

control_mean_scores = list(map(np.mean, get_bootstrap_samples(control.n_clicks.values, 500)))
exp_mean_scores = list(map(np.mean, get_bootstrap_samples(exp.n_clicks.values, 500)))

print("95% Средный доверительный интервал для contol:", stat_intervals(control_mean_scores, 0.05))
print("95% Средный доверительный интервал для exp:", stat_intervals(exp_mean_scores, 0.05))

delta_mean = list(map(lambda x: x[1] - x[0], zip(control_mean_scores, exp_mean_scores)))
print("95% разность средных доверительных интервалов:",  stat_intervals(delta_mean, 0.05))
```

```python
np.random.seed(0)

control_median_scores = list(map(np.median, get_bootstrap_samples(control.n_clicks.values, 500)))
exp_median_scores = list(map(np.median, get_bootstrap_samples(exp.n_clicks.values, 500)))

print("95% Медианный доверительный интервал для contol:", stat_intervals(control_median_scores, 0.05))
print("95% Медианный доверительный интервал для exp:", stat_intervals(exp_median_scores, 0.05))

delta_median = list(map(lambda x: x[1] - x[0], zip(control_median_scores, exp_median_scores)))
print("95% разность медианный доверительных интервалов:",  stat_intervals(delta_median, 0.05))
```

```python
def get_bootstrap_samples(data, n_samples):
    indices = np.random.randint(0, len(data), (n_samples, len(data)))
    samples = data[indices]
    return samples

bootstrap_exp = get_bootstrap_samples(exp.n_clicks.values, n_samples=500)
bootstrap_cnt = get_bootstrap_samples(control.n_clicks.values, n_samples=500)

medians = np.median(bootstrap_exp, axis=1) - np.median(bootstrap_cnt, axis=1)
means = np.mean(bootstrap_exp, axis=1) - np.mean(bootstrap_cnt, axis=1)

alpha = 0.05
print(np.percentile(medians, [100 * alpha / 2., 100 * (1 - alpha / 2.)]))
print(np.percentile(means, [100 * alpha / 2., 100 * (1 - alpha / 2.)]))
```

```python
scipy.stats.mannwhitneyu(
    df[(df.browser=='Browser #4') & (df.slot=='exp')].n_clicks.values,
    df[(df.browser=='Browser #4') & (df.slot=='control')].n_clicks.values,
    alternative='two-sided')
```

```python
x = df[(df.browser=='Browser #4') & (df.slot=='exp')].n_clicks.values
y = df[(df.browser=='Browser #4') & (df.slot=='control')].n_clicks.values
n1 = len(x)
n2 = len(y)
ranked = scipy.stats.rankdata(np.concatenate((x, y)))
rankx = ranked[0:n1]  # get the x-ranks
u1 = n1*n2 + (n1*(n1+1))/2.0 - np.sum(rankx, axis=0)  # calc U for x
u2 = n1*n2 - u1  # remainder is U for y
T = scipy.stats.tiecorrect(ranked)
if T == 0:    
    raise ValueError('All numbers are identical in amannwhitneyu')
sd = np.sqrt(T * n1 * n2 * (n1+n2+1) / 12.0)
meanrank = n1*n2/2.0 + 0.5 
bigu = max(u1, u2)
z = np.abs(bigu - meanrank)
fact2 = 2.
z = z / sd
p_value = scipy.stats.norm.sf(z) * fact2
```

```python
pvals = []
for i in df.browser.drop_duplicates():
    s, p = scipy.stats.mannwhitneyu(
        df[(df.browser==i) & (df.slot=='exp')].n_clicks.values,
        df[(df.browser==i) & (df.slot=='control')].n_clicks.values,
        alternative='two-sided')
    pvals.append(p)

pvals
```

```python
from statsmodels.stats.multitest import multipletests
multipletests(pvals, alpha = 0.05, method = 'bonferroni')
```
