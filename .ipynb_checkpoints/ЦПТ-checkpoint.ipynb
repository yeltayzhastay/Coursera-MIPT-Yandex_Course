# Импортируем неоходимые библиотеки

```python
import matplotlib.pyplot as plt
import scipy.stats as sts
import numpy as np
%matplotlib inline
```

# Использовать будем бета распределение

```python
a = 0.5
b = 0.5
beta = sts.beta(a, b)
sample = beta.rvs(1000)
```

Рассчитываем дисперсию и среднее
$$ D = \frac{\alpha \beta} {(\alpha + \beta)^2 (\alpha + \beta + 1)} = \frac{0.5 \cdot 0.5} {(0.5 + 0.5)^2 (0.5 + 0.5 + 1)} = 0.125 $$
$$ E = \frac{\alpha} {\alpha + \beta} = \frac{0.5} {0.5 + 0.5} = 0.5 $$

```python
dispersia = a * b / pow((a + b), 2) / (a + b + 1)
mean = a / (a + b)

dispersia, mean
```

# Строим гистограмму выборки

```python
x = np.linspace(0, 1, 100)
plt.hist(sample, density=True)

plt.plot(x, beta.pdf(x))
plt.ylabel('fraction of samples')
plt.xlabel('$x$')
```

# Функция строящая гистограмму распределения выборочных средних

```python
def plot_hist(n):
    sd = np.sqrt(dispersia / (n - 1))

    means = []
    for i in range(1000):
        sample = beta.rvs(n)
        means.append(np.mean(sample))

    plt.hist(means, density=True)

    x = np.linspace(0.5 - 2 * sd, 0.5 + 2 * sd, 100)
    pdf = sts.norm(mean, sd).pdf(x)
    plt.plot(x, pdf)
    plt.ylabel('$f(x)$')
    plt.xlabel('$x$')
```

## Строим гистограмму распределения выборочных средних при размере выборки равном 5

$$ sd = \sqrt{\frac{D}{n-1}} = \sqrt{\frac{0.125}{4}} = 0.177 $$

```python
plot_hist(5)
```

## Строим гистограмму распределения выборочных средних при размере выборки равном 10

$$ sd = \sqrt{\frac{D}{n-1}} = \sqrt{\frac{0.125}{9}} = 0.118 $$

```python
plot_hist(10)
```

## Строим гистограмму распределения выборочных средних при размере выборки равном 50

$$ sd = \sqrt{\frac{D}{n-1}} = \sqrt{\frac{0.125}{49}} = 0.051 $$

```python
plot_hist(50)
```

## Строим гистограмму распределения выборочных средних при размере выборки равном 50

$$ sd = \sqrt{\frac{D}{n-1}} = \sqrt{\frac{0.125}{99}} = 0.036 $$

```python
plot_hist(100)
```

## Строим гистограмму распределения выборочных средних при размере выборки равном 50

$$ sd = \sqrt{\frac{D}{n-1}} = \sqrt{\frac{0.125}{999}} = 0.011 $$

```python
plot_hist(1000)
```

# Вывод
При увеличении числа элементов в выборке распределение выборочных средних сжимается около среднего функции распределения.
Чем больше элементов в выборке, тем выше точность аппроксимации распределения выборочных средних нормальным распределением.
