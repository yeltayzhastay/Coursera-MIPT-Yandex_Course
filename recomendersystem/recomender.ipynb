```python
import pandas as pd
import numpy as np
```

```python
def precision(actual, predicted, k):
    act_set = set(actual)
    pred_set = set(predicted[:k])
    result = len(act_set & pred_set) / float(k)
    return result

def recall(actual, predicted, k):
    act_set = set(actual)
    pred_set = set(predicted[:k])
    result = len(act_set & pred_set) / float(len(act_set))
    return result
```

```python
def countering(arr, all_s):
    res = {}
    for i in list(set(arr)):
        res[i] = all_s[i]
    return list({k: v for k, v in sorted(res.items(), key=lambda item: item[1], reverse=True)}.keys())
```

```python
train = pd.read_csv('coursera_sessions_train.txt', sep=';', header=None)
train.columns = ['views', 'purchased']
train.views = train.views.apply(lambda x: [int(i) for i in x.split(',')])
train.purchased = train.purchased.apply(lambda x: [int(i) for i in x.split(',')] if type(x) == str else [])
train
```

```python
v = []
for i in train.views:
    v += i
v = Counter(v)

p = []
for i in train.purchased:
    p += i
p = Counter(p)
```

```python
trainw = train.iloc[pd.read_csv('coursera_sessions_train.txt', sep=';', header=None).dropna().index]
trainw
```

```python
trainw['v'] = trainw.views.apply(lambda x: countering(x, Counter(p)))
trainw['p'] = trainw.purchased.apply(lambda x: countering(x, Counter(v)))
trainw
```
