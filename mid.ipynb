## Центральная предельная теорема своими руками

```python
# загружаем необходимые библиотеки  

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as sts
%matplotlib inline
```

```python

n = 5 # зададим параметры испытаний в выборке 
p = 0.6 # вероятность положительного исхода 
mu = n*p # теоретическое среднее значение 
DX = n*p*(1-p) # теоретическая дисперсия
mu , DX 
```

```python
# сгенерируем биномиальное распределение
data = np.random.binomial(n, p, size=1000) 
# создадим закон нормального распределения 
norm_dist = sts.norm(mu, np.sqrt(DX)) #
```

```python
# Построем гистоограмму для 5 испытаний
plt.hist(data, bins=n+1, density=True)
plt.ylabel('fraction of samples')
plt.xlabel('$x$')

x = np.linspace(0,n,100)
pdf = norm_dist.pdf(x)
plt.plot(x, pdf, label='theoretical pdf', alpha=0.5)
plt.legend()
plt.ylabel('$f(x)$')
plt.xlabel('$x$')
```
