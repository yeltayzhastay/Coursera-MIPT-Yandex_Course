## Математическое ожидание

```python
def mean(arr): # мат ожидания
    sum = 0.0
    iter = 0.0
    for i in arr:
        sum += i #сумматор
        iter += 1 #количество
    return sum/len(arr) #Арифметическое среднее значение
```

## Дисперсия

```python
def var(arr):
    x_sum = []
    x_mean = mean(arr) #вычислим ариф.ср.значение
    for x in arr:
        x_sum.append((x - x_mean)**2) #X - X мат ожидания
    return mean(x_sum)# мат ожидания
```
